## `1.A Dari mana diambil`

Data diambil dari prestasi siswa dalam pendidikan menengah di dua sekolah Portugis. Atribut data mencakup nilai siswa, demografi, fitur sosial, dan sekolah, dan dikumpulkan dengan menggunakan laporan sekolah dan kuesioner.

Data disediakan untuk dua subjek yang berbeda:
*   Matematika `(mat)`
*   Bahasa Portugis `(por)`

Sumber data ini adalah dari paper yang ditulis oleh **[Cortez and Silva, 2008]
**

http://archive.ics.uci.edu/dataset/320/student+performance

## `1.B Bagaimana proses pengambilan datanya`

data dalam bentuk csv, kemudian diupload di github agar bisa diakses. 

```python
# Import Libbrary yang dibutuhkan
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from time import time
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix, roc_curve, accuracy_score, f1_score, roc_auc_score, classification_report
from astropy.table import Table
from sklearn.metrics import roc_auc_score

url_mat = "https://raw.githubusercontent.com/k4ilham/machineLearning/main/hi/student-mat.csv"
url_por = "https://raw.githubusercontent.com/k4ilham/machineLearning/main/hi/student-por.csv"

# Membaca data CSV
mat = pd.read_csv(url_mat, sep=";")
por = pd.read_csv(url_por, sep=";")

# Data Frame Mat & Por
df = pd.concat([mat, por], ignore_index=True)

df.describe()
```

```python
# convert final_score to categorical variable # Good:15~20 Fair:10~14 Poor:0~9
df['final_grade'] = 'na'
df.loc[(df.final_score >= 16) & (df.final_score <= 20), 'final_grade'] = 'good'
df.loc[(df.final_score >= 12) & (df.final_score <= 15), 'final_grade'] = 'fair'
df.loc[(df.final_score >= 0) & (df.final_score <= 11), 'final_grade'] = 'poor'
df.head(5)
```

Ya, terdapat perbedaan antara Train-Test Split dan 10-Fold Cross Validation dalam hal cara pembagian dataset dan evaluasi model. Berikut adalah perbedaan utama keduanya:

1. **Pembagian Dataset:**
   - **Train-Test Split:** Membagi dataset menjadi dua bagian: satu untuk pelatihan (training) dan satu untuk pengujian (testing). Biasanya, sebagian besar data digunakan untuk pelatihan, dan sebagian kecil digunakan untuk pengujian (misalnya, 80% pelatihan, 20% pengujian).
   - **10-Fold Cross Validation:** Membagi dataset menjadi 10 bagian (fold). Model dilatih dan diuji sebanyak 10 kali. Pada setiap iterasi, satu fold digunakan sebagai data pengujian, sedangkan sembilan fold lainnya digunakan sebagai data pelatihan. Proses ini diulang sebanyak 10 kali, sehingga setiap fold menjadi data pengujian tepat satu kali.

2. **Evaluasi Model:**
   - **Train-Test Split:** Model dievaluasi hanya pada satu set pengujian yang telah ditentukan sebelumnya. Evaluasi ini dapat bervariasi tergantung pada bagaimana data dipecah.
   - **10-Fold Cross Validation:** Model dievaluasi pada 10 set pengujian yang berbeda, dan hasil evaluasi akhir adalah rata-rata dari hasil evaluasi pada setiap iterasi. Ini memberikan gambaran yang lebih baik tentang kinerja model di berbagai subset data.

3. **Keuntungan dan Kekurangan:**
   - **Train-Test Split:** Lebih cepat, tetapi hasil evaluasinya dapat bervariasi tergantung pada bagaimana dataset dibagi. Performa model mungkin mendapatkan hasil yang sangat baik atau buruk tergantung pada subset tertentu yang digunakan untuk pengujian.
   - **10-Fold Cross Validation:** Memberikan estimasi kinerja model yang lebih stabil dan dapat diandalkan karena model dievaluasi pada multiple subsets data. Namun, memerlukan lebih banyak waktu komputasi.

Pemilihan antara Train-Test Split dan Cross Validation bergantung pada jumlah data yang tersedia, kebutuhan akan estimasi yang lebih stabil, dan seberapa besar kita ingin melatih model pada seluruh dataset. Cross-validation lebih disukai ketika dataset relatif kecil dan kita ingin meminimalkan efek kebetulan dalam pembagian data.

### `3.B.3 (Bonus) Bandingkan jika menggunakan grid search. Apakah hasil yang diperoleh sama untuk hyperparameternya? Mengapa?`

Apakah hasil yang diperoleh sama untuk hyperparameternya? Kemungkinan Sama: Jika mendefinisikan grid parameter yang sama seperti eksperimen awal  dan menggunakan kriteria seleksi yang sama (misalnya, akurasi), kemungkinan besar  akan mendapatkan hasil yang sama.

Kemungkinan Berbeda: Jika  memperluas atau mengubah grid pencarian, atau jika  menggunakan kriteria seleksi yang berbeda (misalnya, f1-score alih-alih akurasi),  mungkin mendapatkan set hyperparameter yang berbeda.

Mengapa Hasilnya Bisa Berbeda? Ada beberapa alasan mengapa hasil dari GridSearchCV mungkin berbeda dari eksperimen awal , termasuk:

Rentang Pencarian: Jika memperluas rentang pencarian atau menambahkan lebih banyak opsi ke grid , GridSearchCV memiliki lebih banyak kombinasi untuk dipertimbangkan.

Kriteria Seleksi: Seperti yang disebutkan sebelumnya, jika  memilih hyperparameter berdasarkan metrik yang berbeda dari eksperimen awal, mungkin mendapatkan hasil yang berbeda.

Variabilitas dalam Data: Jika  menggunakan validasi silang dengan pembagian yang berbeda atau data  memiliki variabilitas yang tinggi, a mungkin mendapatkan hasil yang berbeda pada setiap kali menjalankan GridSearchCV.

Kompleksitas Model: Untuk model yang lebih kompleks dengan banyak hyperparameter, ada kemungkinan terdapat beberapa set hyperparameter yang memberikan kinerja yang hampir sama. Dalam kasus seperti ini, hasil dari setiap kali menjalankan GridSearchCV mungkin berbeda.

Secara keseluruhan, jika Anda mendapatkan hasil yang berbeda dari GridSearchCV dibandingkan dengan eksperimen awal, penting untuk memahami alasan perbedaan tersebut dan mempertimbangkan hasil dari kedua pendekatan saat memilih hyperparameter terbaik untuk model Anda.
